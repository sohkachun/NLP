# Features selection and preprocess data

```python
import re, contractions
from string import punctuation
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

lemmatizer = WordNetLemmatizer()
stopwords = set(nltk.corpus.stopwords.words('english'))
wordnet_map = {
    'N': wordnet.NOUN,
    'V': wordnet.VERB,
    'J': wordnet.ADJ,
    'R': wordnet.ADV
}

def preprocess_text(text):
  #clean empty space
  text.lower()
  text = re.sub(r'\n|\r', ' ', text)
  #clean html
  cleaner = re.compile('<.*?:')
  text = re.sub(cleaner, ' ', text)
  text = re.sub(r'^a-zA-Z\s', ' ',text, flags=re.I)
  # Using regular expressions to remove "id" followed by any number
  text = re.sub(r"id \d+(?: name)?", "", text)
  #expand contractions
  import contractions
  text = contractions.fix(text)
  #strip punctuation
  from string import punctuation
  text = ''.join(c for c in text if not c in punctuation)+ ' '
  #lemmatize
  lemmatizer = WordNetLemmatizer()
  pos_tagged_text = nltk.pos_tag(text.split())
  tokens = [lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text]
  return ' '.join(tokens)
```

```python
features = ['genres', 'keywords', 'original_language', 'overview', 'tagline']
for feature in features:
    #process each feature
    movies[feature] = movies[feature].astype(str).apply(preprocess_text)
```

```python
movies['concatenated_features'] = movies[features].astype(str).agg(' '.join, axis=1) #concatenate features into a string
```

```python
#create indices for the title
indices = pd.Series(movies.index, index=movies['title'])
indices = indices.drop_duplicates()

indices.head()
```

# Cosine Similarity 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def recommend(title,features):

    #TF-IDF vectorize features
    tfidf = TfidfVectorizer()
    tfidf_matrix = tfidf.fit_transform(features)

    # Calculate cosine similarity between movies based on the 'tfidf_matrix'
    cosine_sim_tfidf = cosine_similarity(tfidf_matrix, tfidf_matrix)

    # Determine the index of the movie with the given 'title'
    movie_index = indices[title]

    # Generate a list of similarity scores for the movie with the given 'title'
    similarity_scores = list(enumerate(cosine_sim_tfidf[movie_index]))

    # Sort similarity scores in descending order, excluding the first entry (movie itself)
    sorted_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:]

    # Select the next 10 most similar movies
    movie_indices = [index for index, _ in sorted_scores[:5]]

    # Construct the recommendation message
    recommend = f"The Five movies that are similar to {title} are:\n"
    for movie_title in list(movies['title'].iloc[movie_indices]):
        recommend += f"{movie_title}\n"

    return recommend
```

```python
print(recommend('Star Wars',movies['concatenated_features']))
```
